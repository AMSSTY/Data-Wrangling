```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = [2.8, 7.3, 9.6, 8.9, 25.1, 11.4]
x = np.asarray(x)
```

```python
ax = plt.figure(figsize=(15,10))

plt.subplot(2,2,1).plot(x , 'r-*')
plt.subplot(2,2,2).plot(x**x , 'g--')
plt.subplot(2,2,3).hist(x,color='c')
plt.subplot(2,2,4).plot(x/x*2,'y-')


plt.savefig('subplot.png')
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1);
plt.plot(x**2 , color='b' , label='x**2');
plt.plot(x*2 , color='r' , label='x**2');
plt.grid()
plt.plot(x**3 , color='g' , label='x**2');
plt.legend()

plt.subplot(2,2,2);
plt.plot(x**2 , color='c' , label='x**2');
plt.legend()
plt.grid()
plt.savefig('grid_plot.png')
```

```python
explode = np.ones(len(x))*.05
plt.pie(x , explode=explode ,radius=1.4, labels=explode);
plt.savefig('pie.png')
```

```python
plt.bar(x , 1 , color='r')
plt.grid()
```
