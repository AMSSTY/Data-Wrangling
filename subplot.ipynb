```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = [2.8, 7.3, 9.6, 8.9, 25.1, 11.4, 6.7, 5.8, 5.5, 6.7, 6.2, 30.4, 9.0, 8.2, 7.6, 8.5, 6.7]
x = np.asarray(x)
```

```python
plt.subplot(2,3,1).plot(x)
plt.subplot(2,3,2).plot(x**x)
plt.subplot(2,3,3).hist(x)
plt.subplot(2,3,4).plot(x/x*2)
plt.subplot(2,3,5).plot(x/5)
plt.subplot(2,3,6).plot(x/x**x);
```

```python
plt.subplot(2,2,1);
plt.plot(x**2 , color='b');
plt.subplot(2,2,2);
plt.plot(x**2 , color='r');

plt.subplot(2,2,3);
plt.plot(x**2 , color='g');

plt.subplot(2,2,4);
plt.plot(x**2 , color='c');
```
